```python
import time
start = time.perf_counter()
import numpy as np
A = np.array([[1, 9], [3, 2]])
B = np.array([[5, 8], [6, 7]])
C = np.empty((2,2))
D = np.empty((2,2))
for i in range(0,2):
  for j in range(0,2):
    C[i][j] = A[i][j] - B[i][j]
    D[i][j] = A[i][j] + B[i][j]

print(C)
print(D)
stop = time.perf_counter()
print("Elasped time : ", stop-start)
```
